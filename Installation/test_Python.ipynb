```python
import igraph as ig
import pandas as pd
```

# Plot a network

```python
# Create the edge list
edge_list = [("Alice", "Bob"), ("Alice", "Charlie"), ("Alice", "Frank"), ("Bob", "Charlie"), ("Bob", "David"),
             ("Charlie", "David"), ("Charlie", "Eve"), ("Charlie", "Grace"), ("David", "Grace"), ("David", "Henry"),
             ("Eve", "Frank"), ("Eve", "Grace"), ("Frank", "Henry"), ("Grace", "Ivy"), ("Jack", "Alice"),
             ("Jack", "Bob"), ("Jack", "David"), ("Jack", "Eve"), ("Jack", "Henry")]
df = pd.DataFrame(edge_list, columns=["source", "target"])

# Create the network (undirected)
g = ig.Graph.DataFrame(df, directed=False, use_vids=False)

# Plot network
ig.plot(g, bbox=(300, 300), vertex_label=g.vs["name"], vertex_color="lightblue")
```

# Calculate network measures

```python
print("Number of nodes:", g.vcount())
print("Number of edges:", g.ecount())
print("Diameter:", g.diameter())
print("Assortativity:", g.assortativity_degree())
print("Edge Density:", g.density())
print("Transitivity (average):", g.transitivity_avglocal_undirected(mode="zero"))
print("Transitivity:", g.transitivity_undirected())
```
